```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```
# load data
data = load_iris()
```

```
num_sample = len(data.target)
```

```
X, y = load_iris(return_X_y=True)
```

```
X_tr, X_te, y_tr, y_te = train_test_split(X, y , test_size = 0.4, random_state = 0)
```

```
print( f' the num of training data is {len(X_tr)}')
print(f' the num of test data is {len(X_te)}')
```

```
# LR 
model_lr = LogisticRegression()
model_lr.fit(X_tr, y_tr)
y_lr_out = model_lr.predict(X_te)
```

```
y_lr_out
```

```
# NB
model_nb = GaussianNB()
model_nb.fit(X_tr, y_tr)
y_nb_out = model_nb.predict(X_te)
```

```
y_nb_out
```

```
acc_nb = accuracy_score(y_te, y_nb_out)
acc_lr = accuracy_score(y_te, y_lr_out)
print(f'accuracy of naive bayes ba is {acc_nb} ')
print(f'accuracy of lostic regression is {acc_lr}')
```

```
# for each class
from sklearn.metrics import classification_report
print(classification_report(y_te, y_lr_out, target_names= data.target_names))
```

```
print(classification_report(y_te, y_nb_out, target_names= data.target_names))
```

```
# try another model
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
model_lda = LinearDiscriminantAnalysis()
model_lda.fit(X_tr, y_tr)
y_lda_out = model_lda.predict(X_te)
```

```
y_lda_out
```

```
acc_lda = accuracy_score(y_te, y_lda_out)
print(f'accuracy of naive bayes ba is {acc_lda} ')
```

```
print(classification_report(y_te, y_lda_out, target_names= data.target_names))
```
